```python
# 선형회귀 - 공부시간과 시험점수 관계
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y = np.array([81,93,91,97])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.01
epochs = 2001

n = len(x1)

for i in range(epochs):
  y_pred = a1 * x1 + a2 * x2 + b
  error = y - y_pred

  a1_diff = (2/n) * sum(-x1 * (error))
  a2_diff = (2/n) * sum(-x2 * (error))
  b_diff = (2/n) * sum(-(error))

  a1 = a1 - lr * a1_diff
  a2 = a2 - lr * a2_diff
  b = b - lr * b_diff

  if i % 100 == 0:
    print("epoch=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f" % (i, a1, a2, b))

print("실제 점수: ", y)
print("예측 점수: ", y_pred)
```

```python
# 다중 선형 회귀
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt

# 공부시간 x와 성적 y의 넘파이 배열
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential()

# 입력층인 동시에 첫번째 은닉층
model.add(Dense(1, input_dim=1, activation='linear'))
model.compile(optimizer='sgd', loss='mse')
model.fit(x,y,epochs=2000)

plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()

hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할 경우 예상 점수는 %.02f점" % (hour, prediction))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt

# 공부시간, 과외시간 x와 성적 y의 넘파이 배열
x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])

model = Sequential()

model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd', loss='mse')
model.fit(x,y,epochs=2000)

hour = 7
private_class = 4
prediction = model.predict([[hour, private_class]])
print("%.f시간을 공부하고 %.f시간의 과외를 받았을 때, 예상 점수는 %.02f점" % (hour, private_class, prediction))
```

```python
# 로지스틱 회귀 - 공부시간과 합격/불합격 여부
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2,4,6,8,10,12,14])  # 공부한 시간
y = np.array([0,0,0,1,1,1,1])     # 합격 유무(1:합격, 0:불합격)

# 모델 실행
# Sequential() 모델 특징
# 입력층 따로 없음, 반드시 첫번째 Dense Layer에 input_dim 나옴
# model.add(Dense(출력, 입력, 활성화 함수))
model = Sequential()
model.add(Dense(1, input_dim=1, activation='sigmoid'))

# 교차 엔트로피 오차 함수를 이용하기 위해 'binary_crossentropy'로 설정
# optimizer(최적화): sgd 경사하강법으로 미세조정
# loss(loss function, cost function 비용함수 >> y-y_pred 최소화)
# binary(이진의) cross entropy(교차 엔트로피)
model.compile(optimizer='sgd', loss='binary_crossentropy')

# 학습
model.fit(x, y, epochs=5000)

# 모델 테스트 데이터 시각화
plt.scatter(x, y) # 실제값(산정도)
plt.plot(x, model.predict(x), 'r')  # 예측값
plt.show()

# 임의의 학습시간 7을 집어넣어 합격 예상 확률 예측
hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할경우, 합격 예상 확률은 %.01f%%입니다." % (hour, prediction * 100))
```
