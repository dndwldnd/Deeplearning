```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2,4,6,8,10,12,14])  # 공부한 시간
y = np.array([0,0,0,1,1,1,1])     # 합격 유무(1:합격, 0:불합격)

# 모델 실행
# Sequential() 모델 특징
# 입력층 따로 없음, 반드시 첫번째 Dense Layer에 input_dim 나옴
# model.add(Dense(출력, 입력, 활성화 함수))
model = Sequential()
model.add(Dense(1, input_dim=1, activation='sigmoid'))

# 교차 엔트로피 오차 함수를 이용하기 위해 'binary_crossentropy'로 설정
# optimizer(최적화): sgd 경사하강법으로 미세조정
# loss(loss function, cost function 비용함수 >> y-y_pred 최소화)
# binary(이진의) cross entropy(교차 엔트로피)
model.compile(optimizer='sgd', loss='binary_crossentropy')

# 학습
model.fit(x, y, epochs=5000)

# 모델 테스트 데이터 시각화
plt.scatter(x, y) # 실제값(산정도)
plt.plot(x, model.predict(x), 'r')  # 예측값
plt.show()

# 임의의 학습시간 7을 집어넣어 합격 예상 확률 예측
hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할경우, 합격 예상 확률은 %.01f%%입니다." % (hour, prediction * 100))
```

```python
# 텐서플로 라이브러리 내 케라스 API 호출
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 깃허브에서 데이터 가져오기
!git clone https://github.com/taehojo/data.git

# 준비된 수술 환자 데이터 불러오기
Data_set = np.loadtxt("./data/ThoraricSurgery3.csv", delimiter=",")

X = Data_set[:, 0:16] # 환자의 진찰 기록
y = Data_set[:, 16]   # 수술 1년 후 사망/생존 여부

# 딥러닝 모델의 구조 결정
model = Sequential()
model.add(Dense(30, input_dim=16, activation='relu')) # 입력층+은닉층 (출력, 입력, 활성화함수)
model.add(Dense(1, activation='sigmoid')) # 출력층 (출력, 활성화함수)

# 딥러닝 모델 실행
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X, y, epochs=500, batch_size=16)
```
