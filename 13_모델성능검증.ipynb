```python
# pandas 라이브러리 불러오기
import pandas as pd

!git clone https://github.com/taehojo/data.git

# 광물 데이터 불러오기
df = pd.read_csv('./data/sonar3.csv', header=None)

df.head()
```

```python
# 데이터 프레임 내에서 분류 클래스의 개수 확인
df[60].value_counts() # 광물:1, 암석:0
```

```python
# 데이터 확인
X = df.iloc[:, 0:60]  # feature들
y = df.iloc[:, 60]    # ground truth (정답)
```

```python
# 모델 생성 - 이진분류(1,0), 노드1, 활성화함수-sigmoid
# 은닉층 2개, 활성화함수-relu, 첫번째 노드 24개, 두번째 노드 10개
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu')) # 입력층인 동시에 첫번째 은닉층
model.add(Dense(10, activation='relu')) # 두번째 은닉층
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 학습
history = model.fit(X, y, epochs=200, batch_size=10)
```

```python
from sklearn.model_selection import train_test_split

# 음파 관련 속성을 X로, 광물의 종류를 y로 저장
X = df.iloc[:, :60] # 피쳐들
y = df.iloc[:, 60]  # 정답

# 학습셋과 테스트셋 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
# 모델 설정
# 입력항목 - 60개, 은닉층의 활성화함수 -'relu'
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))  # 첫번째 은닉층 노드24
model.add(Dense(10, activation='relu'))                # 두번째 은닉층 노드10
model.add(Dense(1, activation='sigmoid'))              # 출력층(1,0)

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
# 모델을 테스트셋에 적용해 정확도 구하기
score = model.evaluate(X_test, y_test)  # 30% 데이터로 성능 측정
print('Test accuracy: ', score[1])
```

```python
# 실습I초음파 광물 예측하기 : k겹 교차 검증
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

import pandas as pd

!git clone https://github.com/taehojo/data.git

# 광물 데이터 불러오기
df = pd.read_csv('./data/sonar3.csv', header=None)

X = df.iloc[:, 0:60]  # 음파 관련 속성
y = df.iloc[:, 60]    # 광물의 종류

k = 5

kfold = KFold(n_splits=k, shuffle=True)
acc_score = []

def model_fn():
  model = Sequential()
  model.add(Dense(24, input_dim=60, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  return model

# 교차 검정
for train_index, test_index in kfold.split(X):
  X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  model = model_fn()
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  history = model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=0)

  accuracy = model.evaluate(X_test, y_test)[1]
  acc_score.append(accuracy)

  avg_acc_score = sum(acc_score) / k

  print('정확도: ', acc_score)
  print('정확도 평균: ', avg_acc_score)
```
